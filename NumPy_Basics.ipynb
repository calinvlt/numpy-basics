Dot product operation using a regular loop

```python
def python_forloop_list_approach(x, w):
  z = 0.
  for i in range(len(x)):
      z += x[i] * w[i]
  return z

a = [1., 2., 3.]
b = [4., 5., 6.]

print(python_forloop_list_approach(a, b))

large_a = list(range(1000))
large_b = list(range(1000))

%timeit python_forloop_list_approach(large_a, large_b)
```

Dot product operation using numpy operation

```python
import numpy as np

def numpy_dotproduct_approach(a, b):
  return a.dot(b)

n_a = np.array(a)
n_b = np.array(b)

print(numpy_dotproduct_approach(np.array(a), np.array(b)))

n_large_a = np.array(large_a)
n_large_b = np.array(large_b)

%timeit numpy_dotproduct_approach(n_large_a, n_large_b)
```
